```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
white_img = np.ones((300, 300, 3), dtype=np.uint8) * 255
cv2.circle(white_img,(180, 200), 25, (36, 142, 255), 20)
cv2.circle(white_img,(100, 200), 25, (140, 219, 103), 20)
cv2.circle(white_img,(140, 130), 25, (251, 38, 67), 20)

cv2.ellipse(white_img,(140, 130),(36,36),60,0,60,(255, 255, 255),-1)
cv2.ellipse(white_img,(100, 200),(36,36),300,0,60,(255, 255, 255),-1)
cv2.ellipse(white_img,(180, 200),(36,36),240,0,60,(255, 255, 255),-1)
# axes were set to 35,35 but some pixels of the circle under the ellipse were not covered so i increased it to 36,36

cv2.putText(white_img, 'OpenCV', (75, 270), cv2.FONT_HERSHEY_DUPLEX, 1, (0, 0, 0), 2)


cv2.imshow('White Image', cv2.cvtColor(white_img, cv2.COLOR_RGB2BGR))
cv2.waitKey(0)
cv2.destroyAllWindows()

# mostly used plt to set the proper coordinates for the image
plt.imshow(white_img)
```

```python
panda = cv2.imread('images/panda.jpg')
capybara = cv2.imread('images/capybara.jpg')

panda_resize = cv2.resize(panda, (400, 400))
capybara_resize = cv2.resize(capybara, (400, 400))

alpha = 0.5
beta = 1 - alpha

manual_img = alpha * panda_resize + beta * capybara_resize
np.clip(manual_img, 0, 255)  #this might not be needed as values would not exceed 0,255 as alpha is between 0 and 1

# the range for cv2.imshow changes for float64 and does not display the image 
# so convert it to uint8 
manual_img = manual_img.astype(np.uint8)
print(manual_img.dtype)

cv2.imshow('manual_blend', manual_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('manual_blend.jpg', manual_img)

```
